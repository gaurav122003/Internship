```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('/content/world_population.csv',  encoding='utf-8')
df.head()
```

```python
df_final = df.drop(['CCA3','Continent'], axis=1)
```

```python
df_final
```

```python
algeria_df = df_final[df_final['Country/Territory'] == 'Algeria']
```

```python
algeria_df
```

```python
years = ['1970', '1980', '1990', '2000', '2010', '2015', '2020', '2022']
populations = [algeria_df[f'{year} Population'].values[0] for year in years]
```

```python
populations
```

```python
plt.figure(figsize=(5, 3))
plt.bar(years, populations, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of { "Algeria" } Over the Years')
plt.show()
```
